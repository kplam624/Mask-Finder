```python
#adding dependencies
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import requests
import pandas as pd
```

```python
#spilinter below

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#adding url 

#DATASOURCE CREDIT: Worldmeters.info

url = 'https://www.worldometers.info/coronavirus/ '

browser.visit(url)
```

```python
#html object 
html = browser.html

#boutifulsoup
soup = BeautifulSoup(html, 'html.parser')
```

```python
table = soup.find('table', attrs= {"class" : "table table-bordered table-hover main_table_countries dataTable no-footer"})
print(table)
table_ro = table.find_all('tr')
print(table_ro)

x = []

for tr in table_ro:
    
    td = tr.find_all('td')
    row = [tr.text for tr in td]
    x.append(row)
```

```python
corono_df = pd.DataFrame(x[1:], columns = ['#', 'Country,Other', 'TotalCases', 'NewCases', 'TotalDeaths',
       'NewDeaths', 'TotalRecovered', 'ActiveCases', 'Serious,Critical',
       'Tot Cases/1M pop', 'Deaths/1M pop', 'TotalTests', 'Tests/\n1M pop\n',
       'Population', 'Continent', 'A', 'B', 'C', 'D'])
corono_df.head(30)
```

```python
new_corona_df = corono_df.replace('\n', '', regex = True)
new_corona_df
```

```python
#checking to see the last part of dataset
new_corona_df.tail(10)
```

```python
#dropping columns that are unecessa

my_list = ['Country,Other', 'TotalCases', 'NewCases', 'TotalDeaths',
       'NewDeaths', 'TotalRecovered', 'ActiveCases', 'Serious,Critical',
       'Tot Cases/1M pop', 'Deaths/1M pop', 'TotalTests', 'Tests/\n1M pop\n',
       'Population', 'Continent']
#reindexing
dropping_df = new_corona_df.reindex(columns = my_list)
dropping_df.head(30)
```

```python
#dropping rows 
#only including countries and continents for final dataset.
final_df = dropping_df.drop(dropping_df.index[[0, 229, 230, 231, 232, 233, 234, 235, 236]])
```
